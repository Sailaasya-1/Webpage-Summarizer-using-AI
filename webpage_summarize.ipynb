The system_prompt sets the rules, context and behvaiour.

```python
system_prompt_leader = "You are a leader who just quicks go through the content and just provide the details why it was the best as well."
```

```python
system_prompt_ux = "You are a digital marketing and UX expert specializing in optimizing websites for engagement and conversions. Review popular websites and explain what design and content strategies make them effective in attracting and retaining users. Suggest actionable improvements that align with modern marketing best practices."
```

```python
print(user_prompt_for(web_development))
```

```python
message1 = [
    {"role": "system", "content": "You are a lazy assistant"},
    {"role": "user", "content": "What is 2 + 5?"}
]
```

```python
MODEL = "llama3.2"
```

```python
response = client.chat.completions.create(model= MODEL, messages=message1)
print(response.choices[0].message.content)
```

```python
def message_to_pass(website):
    return [ {"role" :"system", "content" :system_prompt_leader},
             {"role" : "user", "content": user_prompt_for(website)}]
```

```python
def message_to_pass_ux(website):
    return [ {"role" :"system", "content" :system_prompt_ux},
             {"role" : "user", "content": user_prompt_for(website)}]
```

```python
def summarize_leader(url):
    website = Website(url)
    response = client.chat.completions.create(
        model = MODEL,
        messages = message_to_pass(website)
    )
    return response.choices[0].message.content

def summarize_ux(url):
    website = Website(url)
    response = client.chat.completions.create(
        model = MODEL,
        messages = message_to_pass_ux(website)
    )
    return response.choices[0].message.content
```

```python
def display_summary_ux(url):
    summary = summarize_ux(url)
    display(Markdown(summary))
```

```python
def display_summary_leader(url):
    summary = summarize_leader(url)
    display(Markdown(summary))
```
